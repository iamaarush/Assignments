```python
# Question 1

Matplotlib is a comprehensive library for creating static , animated and interactive visulaizations in python. 
Matplotlib makes easy things easy and hard things possible create publication quality plots make interactive figures
that can zoom,pan ,update .customize visual style and layout.

The five plots that can be plotted using pyplot module of matplotlib are :-

1. plot(x,y)
2. scatter(x,y)
3. bar(x,y)
4. hist(x)
5. pie(x)
```

```python
# Question 2

# Scatter plot is a visual representation of how two variables releate to each other.

# Given data

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as mtp

mtp.scatter(x,y)
mtp.xlabel('X Axis')
mtp.ylabel('Y Axis')
mtp.title('Generated Data')
mtp.show
```

```python
# Question 3

"""The subplot function takes three arguments that describe the layout of the figure .The layout is organized in rows and 
columns which are represented by the first and second argument. The third argument represents the index of the current plot."""

# Given data

# For line 1: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])

mtp.subplot(2,2,1)
mtp.plot(x,y)

# For line 2:
x = np.array([0, 1, 2, 3, 4, 5])  
y = np.array([50, 20, 40, 20, 60, 70])

mtp.subplot(2,2,2)
mtp.plot(x,y)

# For line 3:
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])

mtp.subplot(2,2,3)
mtp.plot(x,y)

# For line 4:
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])

mtp.subplot(2,2,4)
mtp.plot(x,y)

mtp.show
```

```python
# Question 4

"""Bar plots are a type of data visualization used to represent data in the form of rectangular bars. the hieght of each
bar represents the value of data point and the width of each bar represents the category of the data."""

# Given Data

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```

```python
import matplotlib.pyplot as plt

# Bar plot
plt.figure(figsize=(5,3))
plt.bar(company , profit  )
plt.show

```

```python
# Horizontal Bar plot

plt.figure(figsize=(5,3))
plt.barh(company , profit)
plt.show
```

```python
# Question 5

"""A box plot is also known as whisker plot is created to display the summary of the set of data values having properties 
like minimum , first quartile , median , third quartile and maximum. in the box plot a box is created from the first
quartile to the third quartile ,a vertical line is also there which goes through the box at the median. Here x axis
denotes the data to be plotted while the y axis shows the frequency distribution. 
It is also useful in comparing the distribution of data across data sets by drawing boxplots for each of them. """


# Given data 

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
# For box 1

plt.figure(figsize=(5,3))
plt.boxplot(box1)

plt.show
```

```python
# For box 2

plt.figure(figsize=(5,3))
plt.boxplot(box2)
plt.show
```
